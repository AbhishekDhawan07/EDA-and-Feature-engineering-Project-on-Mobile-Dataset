```python
# Outliers Detection and Removing Outliers.

# Detect Outliers Using IQR.
Q1 = df['battery_power'].quantile(0.25)  # 25th percentile
Q3 = df['battery_power'].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Any value below lower_bound or above upper_bound is considered an Outlier.

# Remove Outliers.

# Filter the dataset to remove Outliers.

df_clean = df[(df['battery_power'] >= lower_bound) & (df['battery_power'] <= upper_bound)]

print("Original dataset size:", df.shape)
print("Cleaned dataset size:", df_clean.shape)

# df_clean  dataset without outliers in battery_power.

# You can now use df_clean for analysis or modeling.


import seaborn as sns
import matplotlib.pyplot as plt

# Before removing outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df['battery_power'])
plt.title("Battery Power - Before Removing Outliers")
plt.show()

# After removing outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df_clean['battery_power'])
plt.title("Battery Power - After Removing Outliers")
plt.show()


# It is Clear from the Box - Plot , there are no Outliers in battery_power Column.
```

## Feature Engineering on this Dataset.

```python
#) Scaling / Normalization.

#1) Scaling (or Normalization) is a way of resizing your numbers so that all your features (columns) are on a similar scale.

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['battery_power_scaled'] = scaler.fit_transform(df[['battery_power']])
print(df['battery_power_scaled'])
```

```python
# Binning / Discretization.

# Convert a Numerical Variable into Categories(bins). Useful for features like age, battery, price.

# Binning battery_power into 3 categories: Low, Medium, High.

df['battery_power_bin'] = pd.cut(df['battery_power'], bins=3, labels=['Low','Medium','High'])
print(df['battery_power_bin'])
```
